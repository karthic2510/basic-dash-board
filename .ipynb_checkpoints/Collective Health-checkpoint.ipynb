# Analysing and Visualizing User Data

This notebook was part of a fun assignment I did to test my visualization and basic data wrangling skills. I used the following packages:

* Pandas and Numpy for data wrangling
* Seaborn and Matplotlib for visualization (Seaborn for actual graphs and pyplot, gridspec for refining the graphs)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
import numpy as np
```

## The Data

The data provided by my mentor included two files. One providing information an actual usage statistics, and the other providing information about the account. 

```python
data = pd.read_excel('Product Scorecard Metrics and ED Visits.xlsx')
hos_data = pd.read_csv('hospital data.csv')
```

```python
data.info()
```

```python
data.isnull().sum()
```

***
We begin with the following steps:

1. Create a column to identfy accounts as being active or inactive (This was a key question in the exercise)
2. Create a ratio to identify percentage of users who are active.
3. Define measurement period
4. Replace NAs with 0
5. Fix data types and ensure missing is categorised as such
6. Categorise accounts with missing engagement levels as red. 

```python
data.loc[(data['Facility Last Login(Days)'] != 0), 'Active'] = 'Active'
data.loc[(data['Facility Last Login(Days)'] == 0), 'Active'] = 'Inactive'

data.loc[ (data['Active Users'] != 0) & (data['Total Users'] != 0),'active_total_ratio'] = data['Active Users'] / data['Total Users']
data.loc[ (data['Active Users'] == 0) | (data['Total Users'] == 0),'active_total_ratio'] = 0

data['Measurement Period'] = (data['Effective End'].dt.day - data['Effective Start'].dt.day)

#Creating index of columns with null values
null_col = hos_data.isnull().sum().sort_values(ascending = False)
null_col = null_col[ null_col > 0]
#Looping through to replace NAs with mean values
null_col = null_col.index
for col in null_col:
    hos_data.loc[:, col].fillna('0', inplace = True)
    
#Creating index of columns with null values
null_col = data.isnull().sum().sort_values(ascending = False)
null_col = null_col[ null_col > 0]
#Looping through to replace NAs with mean values
null_col = null_col.index
for col in null_col:
    data.loc[:, col].fillna(0, inplace = True)

hos_data['staffed_beds'] = hos_data['staffed_beds'].astype(int)
hos_data['number_discharges'] = hos_data['number_discharges'].astype(int)
hos_data['est_ED_visits'] = hos_data['est_ED_visits'].astype(int)
hos_data['average_LOS'] = hos_data['average_LOS'].astype(int)
hos_data['LWBS_rate'] = hos_data['LWBS_rate'].astype(int)

for col in ['join_date', 'region', 'geographic_classification', 'hospital_type', 'firm_type']:
    hos_data[col] = hos_data[col].apply(lambda x: 'missing' if x == '0' else x)
    
data['Engagement Level'] = data['Engagement Level'].apply(lambda x: 'Red' if x == 0 else x)
```

```python
data_login = data.query('''`Facility Last Login(Days)` != 0''')
```

### Simplifying numerical data

For visualization purposes, and to answer specific questions, I categorized most numerical data into bins. With more context and knowledge of the business, this can be done based on cut-offs and identified values.

```python
labels_4b = ['very low', 'low', 'medium', 'high']
labels_3b = ['low', 'medium', 'high']
labels_2b = ['low', 'high']
```

```python
%%capture [--no-stderr]
data_login['Login Level'], login_bins = pd.qcut(data_login['Facility Last Login(Days)'], q = [0, 0.25, 0.5, 0.75, 1], retbins= True, duplicates='drop', labels = labels_4b)
login_bins = np.concatenate(([-np.inf], login_bins[1:-1], [np.inf]))
data_login['Login Level'].value_counts()
```

```python
%%capture [--no-stderr]
data_login['Login Ratio Level'], login_ratio_bins = pd.qcut(data_login['Ratio Days Logged In'], q = [0, 0.25, 0.5, 0.75, 1], retbins= True, duplicates='drop', labels = labels_4b)
login_ratio_bins = np.concatenate(([-np.inf], login_ratio_bins[1:-1], [np.inf]))
data_login['Login Ratio Level'].value_counts()
```

```python
%%capture [--no-stderr]
data_login['Active User Level'], active_ratio_bins = pd.qcut(data_login['active_total_ratio'], q = [0, 0.25, 0.5, 0.75, 1], retbins= True, duplicates='drop', labels = labels_4b)
active_ratio_bins = np.concatenate(([-np.inf], active_ratio_bins[1:-1], [np.inf]))
data_login['Active User Level'].value_counts()
```

```python
%%capture [--no-stderr]
data_login['Time Per User Level'], time_user_bins = pd.qcut(data_login['Avg Time per User'], q = [0, 0.25, 0.5, 0.75, 1], retbins= True, duplicates='drop', labels = labels_4b)
time_user_bins = np.concatenate(([-np.inf], time_user_bins[1:-1], [np.inf]))
data_login['Time Per User Level'].value_counts()
```

```python
%%capture [--no-stderr]
data_login['Notification Levels'], notifications_bins = pd.qcut(data_login['Notifications'], q = [0, 0.25, 0.5, 0.75, 1], retbins= True, duplicates='drop', labels = labels_4b)
notification_bins = np.concatenate(([-np.inf], notifications_bins[1:-1], [np.inf]))
data_login['Notification Levels'].value_counts()
```

```python
data['Notification Level'] = pd.cut(data['Notifications'], bins = notification_bins, labels = labels_4b)
data['Time Per User Level'] = pd.cut(data['Avg Time per User'], bins = time_user_bins, labels = labels_4b)
data['Active User Level'] = pd.cut(data['active_total_ratio'], bins = active_ratio_bins, labels = labels_4b)
data['Login Ratio Level'] = pd.cut(data['Ratio Days Logged In'], bins = login_ratio_bins, labels = labels_4b)
data['Login Level'] = pd.cut(data['Facility Last Login(Days)'], bins = login_bins, labels = labels_4b)
data_nologin = data.query('''`Facility Last Login(Days)` == 0''')
```

```python
data_comp = data[['id', 'Total Users', 'Engagement Level', 'DAU/MAU', 'Active', 'Notification Level', 'Time Per User Level', 'Active User Level', 'Login Ratio Level', 'Login Level']]
data_comp = data_comp.merge(hos_data, on = 'id', how = 'left')
data_comp.info()

#Creating index of columns with null values
null_col = data_comp.isnull().sum().sort_values(ascending = False)
null_col = null_col[ null_col > 0]
#Looping through to replace NAs with mean values
null_col = null_col.index
for col in null_col:
    data_comp.loc[:, col].fillna(0, inplace = True)

for col in ['state','join_date', 'region', 'geographic_classification', 'hospital_type', 'firm_type']:
    data_comp[col] = data_comp[col].apply(lambda x: 'missing' if x == 0 else x)
    
data_comp = data_comp.query('''`state` != "missing"''')
data_comp = data_comp.query('''`firm_type` != "missing"''')
```

### The Final Dataset

The final dataset used for visualizations below is the data_comp dataset. This combined both the files provided, created new variables for better visualization and also removed missing data. In reality, we cannot remove missing data, and instead would attempt to fix the data pipeline to ensure all data is capture.

### Where do we focus? 

Based on examination of active and inactive accounts, a few clear trends popped out. However for the second part of the question, I decided to dive deeper into accounts where:

* Engagement level is Red (Worst)
* Proportion of Active Users to Total Users is high. 

Clearly there were many users in these accounts who were active and found utility, however, they had low levels of engagement. 

```python
data_where = data.query('''`Engagement Level` == "Red"''')
data_where = data_where.query('''`Active User Level` == "high"''')
data_where = data_where.merge(hos_data, on = 'id', how = 'left')

#Creating index of columns with null values
null_col = data_where.isnull().sum().sort_values(ascending = False)
null_col = null_col[ null_col > 0]
#Looping through to replace NAs with mean values
null_col = null_col.index
for col in null_col:
    data_where.loc[:, col].fillna(0, inplace = True)

for col in ['state','join_date', 'region', 'geographic_classification', 'hospital_type', 'firm_type']:
    data_where[col] = data_where[col].apply(lambda x: 'missing' if x == 0 else x)
    
data_where = data_where.query('''`state` != "missing"''')
data_where = data_where.query('''`firm_type` != "missing"''')
```

```python
sns.set_context('talk')
fig2 = plt.figure(figsize = (24, 10))
gs.GridSpec(2,2)

plt.subplot2grid((2,2), (0,0), rowspan = 1, colspan = 2)
sns.barplot(data = data_where, x = 'state', y = 'Total Users', estimator = lambda x:len(x),
                order = data_where[['state', 'Total Users']].groupby('state').count().sort_values(by = 'Total Users', ascending = False).index.to_list(),
                ci = False);
plt.xlabel("State")
plt.ylabel("Number of Hospitals")

plt.subplot2grid((2,2), (1, 0), rowspan = 1, colspan = 1)
sns.swarmplot(data = data_where, x = 'staffed_beds', y = 'hospital_type', hue = 'firm_type', size = 5.5)
labs10, ticks10 = plt.yticks()
plt.yticks(labels = ['Critical\nAccess\nHospital', 'Short term\nAcute Care\nHospital', 'Health\nSystem'], ticks = labs10, wrap = True)
plt.xlabel('Number of staffed beds')
plt.ylabel("")
plt.legend(title = '')

plt.subplot2grid((2,2), (1,1))
sns.scatterplot(data = data_where, x = 'Total Users', y = 'DAU/MAU', hue = 'geographic_classification')
plt.xlabel("Number of Total Users")
plt.ylabel("Ratio of Daily to Monthly\nActive Users")
plt.legend(title = '')


plt.tight_layout(pad = 1.0001)
plt.savefig('fig4.png')
_ = plt.show()
```

### Is there a difference based on join date?

```python
query_data['join_date'] = pd.to_datetime(query_data.join_date, format = '%m/%d/%Y')
import datetime
query_data['data_class'] = query_data['join_date'].apply(lambda x: 1 if x < datetime.datetime(2019, 1, 1) else 0)
```

```python
fig3 = plt.figure(figsize = (20, 10));
gs.GridSpec(1,2)

mean_los = query_data.query('''`data_class` == 1''').loc[:, 'average_LOS'].mean()
mean_los2 = query_data.query('''`data_class` == 0''').loc[:, 'average_LOS'].mean()
plt.subplot2grid((1,2), (0,0))
sns.violinplot(data = query_data, y = 'average_LOS', x = 'data_class', cut = 0, order=[1,0]);
ticksv, labsv = plt.xticks()
_ = plt.xticks(labels = ['Before 2019', 'After 2019'], ticks = ticksv)
plt.axhline(y = mean_los, color = 'blue', linestyle = '--', linewidth = 1)
plt.axhline(y = mean_los2, color = 'orange', linestyle = '--', linewidth = 1)
plt.xlabel('Date of Joining')
plt.ylabel('Average Length of Stay in ED')

mean_sb = query_data.query('''`data_class` == 1''').loc[:, 'staffed_beds'].mean()
mean_sb2 = query_data.query('''`data_class` == 0''').loc[:, 'staffed_beds'].mean()
plt.subplot2grid((1,2), (0,1))
sns.violinplot(data = query_data, y = 'staffed_beds', x = 'data_class', order= [1,0])
plt.ylim(0, 1500)
ticksv, labsv = plt.xticks()
_ = plt.xticks(labels = ['Before 2019', 'After 2019'], ticks = ticksv)
plt.axhline(y = mean_sb, color = 'blue', linestyle = '--', linewidth = 1)
plt.axhline(y = mean_sb2, color = 'orange', linestyle = '--', linewidth = 1)
plt.xlabel('Date of Joining')
plt.ylabel('Number of staffed beds')

plt.savefig('fig5.png')
_ = plt.show()
```
